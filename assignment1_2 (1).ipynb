```python
import pandas as pd
import matplotlib.pyplot as plt

# Given data
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Creating a DataFrame
healthcare_data = pd.DataFrame({
    'Name': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})

# 3. Scatter Plot

plt.scatter(healthcare_data['Age'], healthcare_data['Weight'], color='blue')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age')
plt.ylabel('Weight')


# 4. Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(healthcare_data['Name'], healthcare_data['Height'], color='green')
plt.title('Bar Chart of Height for Each Person')
plt.xlabel('Name')
plt.ylabel('Height')


# 5. Histogram
plt.figure(figsize=(8, 6))
plt.hist(healthcare_data['Weight'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')

```
